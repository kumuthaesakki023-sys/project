```python
import pandas as pd

data = {
    'Attrition': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'Department': ['Sales', 'HR', 'IT', 'Sales', 'IT'],
    'MonthlyIncome': [30000, 45000, 25000, 50000, 28000],
    'TotalWorkingYears': [2, 5, 1, 8, 3],
    'JobSatisfaction': [2, 4, 1, 3, 2],
    'WorkLifeBalance': [2, 3, 1, 4, 2]
}

df = pd.DataFrame(data)

df.to_csv('HR_Employee_Attrition.csv', index=False)

print("Dataset Created Successfully")
print(df)
```

```python
# Check Missing Values
print(df.isnull().sum())

# Remove Missing Values
df = df.dropna()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("HR_Employee_Attrition.csv")

print(df['Attrition'].value_counts())

df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Employee Attrition'
)
plt.ylabel('')
plt.show()

df[df['Attrition'] == 'Yes']['Department'].value_counts().plot(
    kind='bar',
    title='Department Wise Attrition'
)
plt.xlabel('Department')
plt.ylabel('Employees Left')
plt.show()
```

```python
print("\nJob Satisfaction vs Attrition:")
print(df.groupby('JobSatisfaction')['Attrition'].value_counts())

print("\nWork Life Balance vs Attrition:")
print(df.groupby('WorkLifeBalance')['Attrition'].value_counts())
```
